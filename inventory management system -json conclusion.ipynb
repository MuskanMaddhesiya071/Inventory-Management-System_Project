```python
##  NOW GENERATING SALES STRUCTURE IN A FILE

# importing the lib
import json
import time

# INITIALISING VARIABLES
sale=''


# LOADING THE INVENTORY RECORDS FROM JSON TO STRING
f = open("inventory mgmt sys.json", "r")
js = f.read()
f.close()

# CONVERTING STRING RECORD TO JSON/DICT
record = json.loads(js)


# DISPLAYING OUR MENU
print("--------------------MENU---------------------")
for key in record.keys():
    print(key, record[key]['Name'], record[key]['Price'], record[key]['Qn'])
    print("---------------------------------------------")
    print('')
    
    
# TAKING USER INPUT ABOUT THE PURCHASE AND USER DETAILS
ui_name=input("enter your name      : ")
ui_mail=input("enter your email    : ")
ui_ph=input("enter your phone number: ")

ui_pr = str(input("Enter product ID : "))
ui_qn = int(input("Enter Quantity   : "))


# IF WE HAVE ENOUGH QUANTITY THAT USER HAVE
if record[ui_pr]["Qn"] >= ui_qn:
        
        # DISPLAYING PURCHASE DETAIL
        print("--------------------------------------------")
        print("                BILL                        ")
        print("Name      : ", record[ui_pr]["Name"])
        print("Price (Rs): ", record[ui_pr]["Price"])
        print("Quantity  : ", ui_qn)
        print("---------------------------------------------")
        billing_amount = record[ui_pr]["Price"] * ui_qn
        
        print("Billing Amount   : ", billing_amount, "Rs")
        print("---------------------------------------------")
        print("purchase>=500----->5% discount")
        print("purchase>=1000----->10% discount")
        # UPDATING RECORDS DICT
        record[ui_pr]['Qn'] = record[ui_pr]['Qn'] - ui_qn
        if (billing_amount>=500 and billing_amount<=1000):
            billing_amount=billing_amount-0.05
        elif (billing_amount>1000):
            billing_amount=billing_amount-0.10
        else:
            print(billing_amount)
            
                    
        # GENERATING THE SALES STRUCTURE 
        sale = '1'+","+ui_name+","+ui_mail+","+ui_ph+","+ui_pr+","+record[ui_pr]["Name"]+","+str(ui_qn)+","+str(record[ui_pr]["Price"])+","+str(ui_qn * record[ui_pr]["Price"])+","+time.ctime()+"\n"

# IF WE HAVE NOT ENOUGH QUANTITY        
else:
    
    # DISPLAYING INFO ABOUT THE QUANTITY
    print("Sorry, we do not have enough quantity")
    print("We have only", record[ui_pr]["Qn"], "quantity")
    print("---------------------------------------------")
    
    
    ch = input("Enter y/Y to purchase available stock and press n to leave ")
    
    if ch == "y" or ch == "Y":
        
        # DISPLAYING DETAILS
        print("--------------------------------------------")
        print("                BILL                        ")
        print("Name      : ", record[ui_pr]["Name"])
        print("Price (Rs): ", record[ui_pr]["Price"])
        print("Quantity  : ", record[ui_pr]["Qn"])
        print("---------------------------------------------")
        print("Billing Amount   : ", record[ui_pr]["Price"] * record[ui_pr]["Qn"], "Rs")
        print("---------------------------------------------")
        
        
        record[ui_pr]['Qn'] = 0
        sale =ui_name+","+ui_mail+","+ui_ph+","+ui_pr+","+record[ui_pr]["Name"]+","+str(record[ui_pr]['Qn'])+","+str(record[ui_pr]["Price"])+","+str(record[ui_pr]['Qn'] * record[ui_pr]["Price"])+","+time.ctime()+"\n"

        
        
    else:
        print("Thanks")
# CONVERTING DICT TO STRING
js = json.dumps(record)

# SAVING RECORDS IN JSON FILE
f = open("inventory mgmt sys.json", "w")
f.write(js)
f.close()

# SAVING RECORDS IN SALES FILE
f=open("salesTrack.txt","a")
f.write(sale)
f.close()
```
